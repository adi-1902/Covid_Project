```python
import pandas as pd
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
plt.style.use('ggplot')
import numpy as np
```

```python
df = pd.read_csv(r"C:\Users\SINGHAD1\OneDrive - Alcon\Desktop\PortfolioProjects\AlexData.csv")
```

```python
df.head(5)
```

```python
#Checking for missing data

for col in df.columns:
    miss_pct = np.mean(df[col].isnull())
    print("{} - {}%".format(col,miss_pct))
```

```python
#Data Types

df.dtypes
```

```python
df['rel_years'] = df['released'].astype(str).str.split(pat = ' ').str[2]

df['gross'] = df['gross'].astype('Int64')
```

```python
df.dtypes
```

```python
df.sort_values(by = ['gross'] , inplace=False,ascending = False)

```

```python
df.dropna(inplace= True)
```

```python
#Duplicates

#df['company'] = df['company'].drop_duplicates().sort_values(ascending=False)
```

```python
#Scatter Plot
plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget Vs Gross Earnings')
plt.xlabel('Gross')
plt.ylabel('Budget')
```

```python
sns.regplot(x = 'budget', y= 'gross', data = df, scatter_kws={'color':'Red'}, line_kws={'color':'Blue'})
```

```python
#Correlation

df.corr(method = 'pearson')
```

```python
#Correlation Matrix

corr_martrix = df.corr(method = 'pearson')
```

```python
sns.heatmap(corr_martrix,annot= True)

plt.show()
```

```python
#Numerizing Company column

df_num = df

for col_nam in df_num.columns:
    if(df_num[col_nam].dtype == 'object'):
        df_num[col_nam] = df_num[col_nam].astype('category')
        df_num[col_nam] = df_num[col_nam].cat.codes

df_num
```

```python
corr_martrix = df_num.corr(method = 'pearson')

sns.heatmap(corr_martrix,annot= True)
```

```python
df_num.corr()
```

```python
corr_num_mat = df_num.corr()

corr_pair = corr_num_mat.unstack()

corr_pair
```

```python
sorted_pair = corr_pair.sort_values()

sorted_pair
```

```python
high_corr = sorted_pair[(sorted_pair) > 0.5]
high_corr
```
